```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
import progressbar
import time

def invboxcox(y, lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
namefile = 'WAG_C_M.csv' #Файл с данными
name_column = 'WAG_C_M' #Значения временного ряда
date_column = 'Date' #Колонка с датами
period_freq = 12 #Периодичность данных
delta = 0 #Смещение для отрицательных значений ряда
end_date = '2016-08-01' #Дата, с которой мы предсказываем ряд
format_date = '%Y-%m-%d' #Формат даты
period_forecast = 3 #Количество периодов предсказания
```

```python
#bars = pd.read_csv('bars_calc_dprev.csv', ';')
bars = pd.read_csv(namefile, ';', index_col=[date_column], parse_dates=[date_column], dayfirst=True)
print(bars.head())
```

```python
bars.plot(y = name_column, color = 'blue')
plt.show()
```

Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(bars[name_column], freq = period_freq).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(bars[name_column])[1])
```

### Стабилизация дисперсии

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии:

```python
bars['val_box'], lmbda = stats.boxcox(bars[name_column]+delta)
plt.figure(figsize(15,7))
bars.val_box.plot()
plt.ylabel(u'Transformed bars value')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(bars.val_box)[1])
```

### Стационарность

Критерий Дики-Фуллера отвергает гипотезу нестационарности, но визуально в данных виден тренд. Попробуем сезонное дифференцирование; сделаем на продифференцированном ряде STL-декомпозицию и проверим стационарность:

```python
speriod = period_freq
bars['val_box_diff'] = bars.val_box - bars.val_box.shift(speriod)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(bars.val_box_diff[speriod:], freq=period_freq).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(bars.val_box_diff[speriod:])[1])
```

Критерий Дики-Фуллера не отвергает гипотезу нестационарности, и полностью избавиться от тренда не удалось. Попробуем добавить ещё обычное дифференцирование:

```python
bars['val_box_diff2'] = bars.val_box_diff - bars.val_box_diff.shift(1)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(bars.val_box_diff2[speriod+1:], freq=period_freq).plot()   
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(bars.val_box_diff2[speriod+1:])[1])
```

Гипотеза нестационарности отвергается, и визуально ряд выглядит лучше — тренда больше нет.

### Подбор модели

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(bars.val_box_diff2[speriod+1:].values.squeeze(), lags=50, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(bars.val_box_diff2[speriod+1:].values.squeeze(), lags=50, ax=ax)
pylab.show()
```

Начальные приближения: Q=1, q=1, P=1, p=1

```python
ps = range(0, 3)
d=1
qs = range(0, 3)
Ps = range(0, 2)
D=1
Qs = range(0, 2)
```

```python
parameters_list = list(product(ps, qs, Ps, Qs))
#parameters_list = [(0, 2, 0, 1)]
tn = len(parameters_list)
print(tn)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
bar = progressbar.ProgressBar(maxval=tn).start()

t = 0
for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    #print(param)
    bar.update(t)
    t += 1
    try:
        model=sm.tsa.statespace.SARIMAX(bars.val_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], period_freq)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
bar.finish()
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

Её остатки:

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[speriod+1:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[speriod+1:].values.squeeze(), lags=50, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[speriod+1:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[speriod+1:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой). Посмотрим, насколько хорошо модель описывает данные:

```python
bars['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
bars[name_column].plot()
bars.model[speriod+1:].plot(color='r')
plt.ylabel('Bars value')
pylab.show()
```

### Прогноз

```python
bars2 = bars[[name_column]]
date_list = [datetime.datetime.strptime(end_date, format_date) + relativedelta(months=x) for x in range(0, period_forecast*period_freq)]
future = pd.DataFrame(index=date_list, columns=bars2.columns)
bars2 = pd.concat([bars2, future])
bars2['forecast'] = invboxcox(best_model.predict(start=bars_len, end=bars_len+period_forecast*period_freq), lmbda)
plt.figure(figsize(15,7))
bars2[name_column].plot()
bars2.forecast.plot(color='r')
plt.ylabel('Bars value')
pylab.show()
```
