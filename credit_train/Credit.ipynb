```python
import numpy as np
import bigdata
from keras import backend
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(x_train, y_train) = bigdata.load_data("input.csv", 1)
x_train = np.array(x_train)
y_train = np.array(y_train)
print("x_train = ", x_train)
print("y_train = ", y_train)
```

```python
model = Sequential()
layer1 = Dense(2, input_dim=3, activation="sigmoid", kernel_initializer="normal")
layer2 = Dense(1, activation="sigmoid", kernel_initializer="normal")
model.add(layer1)
model.add(layer2)
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
print(model.summary())
```

```python
history = model.fit(x_train, y_train, batch_size=1, epochs=800, verbose=0)
```

```python
print("layer1") 
print(layer1.get_weights())
print("layer2") 
print(layer2.get_weights())
```

```python
plt.plot(history.history['acc'], label='Аккуратность на обучающем наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Аккуратность')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'], label='Ошибка на обучающем наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Ошибка')
plt.legend()
plt.show()
```
